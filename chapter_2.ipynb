**(h)** Which suburb of Boston has lowest median value of owner- occupied homes? What are the values of the other predictors for that suburb, and how do those values compare to the overall ranges for those predictors? Comment on your findings.

```python
# Find suburb with lowest median home value
min_medv_idx = boston['medv'].idxmin() # index of the minimum medv
min_medv_suburb = boston.loc[min_medv_idx]
#
# Get overall ranges for comparison
ranges = boston.describe()
#
# Print results
print("Suburb with lowest median home value:")
print("-" * 40)
print(f"Median home value: ${min_medv_suburb['medv']*1000:.2f}")
print("\nOther characteristics of this suburb:")
print("-" * 40)
for column in boston.columns:
    value = min_medv_suburb[column]
    min_val = ranges.loc['min', column]
    max_val = ranges.loc['max', column]
    percentile = (value - min_val) / (max_val - min_val) * 100
    print(f"{column}:")
    print(f"  Value: {value:.2f}")
    print(f"  Range: [{min_val:.2f} - {max_val:.2f}]")
    print(f"  Percentile: {percentile:.1f}%\n")
```

In few predictors, lowest median value of owner-occupied corresponds to a extreme side of the range of that predictor. This sugest relationship between those predictors and the median value of owner-occupied homes.
